### **⚡ Exercício 1: Função Linear - Engenharia Elétrica (Lei de Ohm)**  

<br> Em um laboratório de circuitos elétricos, um estudante realiza medições em um resistor de resistência constante (ôhmico). Durante o experimento, observou-se que, ao aplicar uma corrente elétrica de 2 A, a diferença de potencial (tensão) medida nos terminais do componente foi de 20 V. Com base nos conceitos da Lei de Ohm, que estabelece uma relação linear entre tensão ($V$) e corrente ($I$) através da fórmula $V = R \cdot I$, determine: O valor da resistência elétrica ($R$) do componente. A função linear $V(I)$ que descreve o comportamento deste sistema. O valor da tensão gerada quando a corrente que atravessa o resistor for de 10 A. <br>


**✅RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">

<br> Dado que $V = 20$ quando $I = 2$, temos:$$20 = R \cdot 2 \implies R = \frac{20}{2} = 10 \, \Omega$$<br>**Resultado final:** A função que descreve o sistema é $V(I) = 10I$. Para $I = 10A$, a tensão será $V(10) = 10 \cdot 10 = 100V$.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_tensao(I):
    return 10 * I

corrente = np.linspace(0, 12, 100)
tensao = calcular_tensao(corrente)

plt.figure(figsize=(8, 5))
plt.plot(corrente, tensao, label='$V(I) = 10I$', color='blue', linewidth=2)
plt.scatter(10, 100, color='red', zorder=5, label='Ponto Projetado (10A, 100V)')
plt.title('Relação Tensão vs Corrente (Projeto A3)')
plt.xlabel('Corrente (I) [Amperes]')
plt.ylabel('Tensão (V) [Volts]')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

###**⚡ Exercício 2: Função Linear — Engenharia Elétrica (Termorresistência)**

Em sistemas de manutenção preditiva, é fundamental monitorar a temperatura dos enrolamentos de cobre de um motor elétrico, uma vez que a resistência elétrica aumenta linearmente com o calor. Durante um teste de bancada, um engenheiro determinou que a resistência de um enrolamento específico segue a função linear:<br><br>

$R (T) = 0,2 T + 46 $<br> <br>
**Onde**:<br>

R  é a resistência elétrica medida em Ohms ( Ω ). T  é a temperatura do enrolamento medida em Graus Celsius ( ∘C ). Considerando que, em regime de carga máxima, o motor atinge uma temperatura de operação estável de  85∘C , calcule o valor da resistência elétrica do enrolamento nestas condições.

**✅RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">

Para determinar a resistência na temperatura de operação de $85^\circ C$, substituímos $T = 85$ na função encontrada:$$R(85) = 0,2(85) + 46$$$$R(85) = 17 + 46 = 63,0 \, \Omega$$ <br>**Resultado final**: A resistência do enrolamento a $85^\circ C$ será de $63,0 \, \Omega$.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_resistencia(T):
    return 0.2 * T + 46

temperaturas = np.linspace(20, 100, 100)
resistencias = calcular_resistencia(temperaturas)

plt.figure(figsize=(9, 5))
plt.plot(temperaturas, resistencias, color='darkorange', label='$R(T) = 0,2T + 46$', linewidth=2)
plt.scatter(85, 63, color='black', zorder=5, label='Ponto de Operação (85°C, 63$\\Omega$)')
plt.title('Variação da Resistência em Função da Temperatura (Projeto A3)')
plt.xlabel('Temperatura (T) [°C]')
plt.ylabel('Resistência (R) [$\\Omega$]')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

###**🧪 Exercício 3: Função Exponencial - Engenharia Química (Decaimento)**

No início do processo ($t = 0$), a concentração medida foi de $C_0 = 500 \, mg/L$. Após exatas $2$ horas de tratamento contínuo, uma nova análise laboratorial indicou que a concentração foi reduzida para $250 \, mg/L$. Com base nestes dados experimentais, determine o valor da constante de decaimento $k$ e calcule o tempo total necessário, a partir do início do processo, para que a concentração do poluente atinja o limite de segurança ambiental de $50 \, mg/L$. Por fim, utilize o Python para plotar a curva de decaimento e sinalizar graficamente o ponto em que o limite de segurança é atingido.



**✅RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">




Com a função definida como $C(t) = 500 \cdot e^{-0,3466t}$, determinamos o tempo $t$ necessário para atingir $50 \, mg/L$:$$50 = 500 \cdot e^{-0,3466t} \implies 0,1 = e^{-0,3466t}$$$$\ln(0,1) = -0,3466t \implies t = \frac{\ln(0,1)}{-0,3466} \approx 6,64 \, \text{horas}$$**Resultado final:** A constante de decaimento é $k \approx 0,3466$ e o tempo necessário é de aproximadamente $6,64$ horas.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

C0 = 500
k = 0.3466

def concentracao(t):
    return C0 * np.exp(-k * t)

t_espaco = np.linspace(0, 10, 100)
c_valores = concentracao(t_espaco)
plt.figure(figsize=(10, 5))
plt.plot(t_espaco, c_valores, color='seagreen', label='Concentração $C(t)$', linewidth=2)
plt.axhline(50, color='red', linestyle='--', label='Limite de Segurança (50 mg/L)')
plt.scatter(6.64, 50, color='black', zorder=5, label='Tempo Crítico (~6,64h)')
plt.title('Monitoramento de Neutralização Química (Projeto A3)')
plt.xlabel('Tempo (t) [Horas]')
plt.ylabel('Concentração (C) [mg/L]')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

###**🔊 Exercício 4: Função Logarítmica — Engenharia Acústica (Nível de Intensidade Sonora)**

Nesta fórmula, $I$ representa a intensidade física do som (em $W/m^2$) e $I_0 = 10^{-12} \, W/m^2$ é uma constante que representa o limiar da audição humana. Sabendo que as normas de segurança do trabalho estabelecem que o nível de ruído para uma jornada confortável não deve ultrapassar os $85 \, dB$, determine, através de cálculos logarítmicos, qual é a intensidade sonora $I$ máxima permitida nesse ambiente. Por fim, utilize a linguagem Python para plotar um gráfico que demonstre a relação entre a intensidade sonora e o nível em decibéis, destacando o limite de segurança estabelecido.



**✅RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">

Substituição e Simplificação:$$85 = 10 \cdot \log_{10} \left( \frac{I}{10^{-12}} \right) \implies 8,5 = \log_{10} \left( \frac{I}{10^{-12}} \right)$$Conversão para Potência (Definição de Logaritmo):$$10^{8,5} = \frac{I}{10^{-12}}$$Cálculo de $I$ (Produto de Potências):$$I = 10^{8,5} \cdot 10^{-12} \implies I = 10^{-3,5} \, W/m^2
$$**Resultado final**: A intensidade máxima permitida é $I \approx 3,16 \cdot 10^{-4} \, W/m^2$.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt


I0 = 1e-12

def calcular_db(I):
    return 10 * np.log10(I / I0)

intensidades = np.logspace(-12, -2, 500)
decibeis = calcular_db(intensidades)

plt.figure(figsize=(10, 5))
plt.semilogx(intensidades, decibeis, color='BLUE', label='Nível Sonoro (dB)', linewidth=2)
plt.axhline(85, color='red', linestyle='--', label='Limite de Segurança (85 dB)')
plt.scatter(3.16e-4, 85, color='black', zorder=5, label='Ponto Crítico (~3,16e-4 $W/m^2$)')
plt.title('Escala de Intensidade Sonora em Data Centers (Projeto A3)')
plt.xlabel('Intensidade Sonora (I) [$W/m^2$] - Escala Logarítmica')
plt.ylabel('Nível de Som (L) [dB]')
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.legend()
plt.show()
```

###**⚡ Exercício 5: Função Trigonométrica — Engenharia Elétrica (Análise de Corrente Alternada)**


Com base nesse modelo matemático, determine qual é a tensão máxima (amplitude) que o sistema atinge em seus picos e calcule a frequência da oscilação em Hertz ($Hz$), comparando-a com o padrão da rede elétrica brasileira. Além disso, determine o valor exato da tensão no instante de tempo $t = 0,005 \, s$. Por fim, utilize a linguagem Python para plotar o comportamento da tensão durante os primeiros dois ciclos completos ($0 \le t \le 0,033 \, s$), demonstrando a natureza periódica do sinal elétrico.

**✅RESPOSTA**

✏️ **Resolução Analítica**:


<div align="justify">



   Analisar a função de tensão $V(t) = 180 \cdot \sin(120\pi t)$. Tensão Máxima ($V_{max}$):
   
   Identificada diretamente pelo coeficiente da amplitude na função seno:$$V_{max} = 180 \, V$$
   
   
   Frequência ($f$): Sabendo que a velocidade angular é $\omega = 120\pi$ e $\omega = 2\pi f$:$$120\pi = 2\pi f \implies f = \frac{120}{2} = 60 \, Hz$$
   
   Cálculo Instantâneo ($t = 0,005 \, s$): Substituindo $t$ na função:
   
   
   $$V(0,005) = 180 \cdot \sin(120\pi \cdot 0,005) = 180 \cdot \sin(0,6\pi)$$$$V(0,005) \approx 180 \cdot 0,951 \approx 171,19 \, V$$**Resultado Final:** A tensão de pico é $180 \, V$, a frequência é $60 \, Hz$ e a tensão no instante solicitado é $171,19 \, V$.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def tensao(t):
    return 180 * np.sin(120 * np.pi * t)

tempo = np.linspace(0, 0.033, 500)
v_valores = tensao(tempo)


plt.figure(figsize=(9, 4))
plt.plot(tempo, v_valores, color='blue', label='$V(t) = 180 \\cdot \\sin(120\\pi t)$')
plt.axhline(0, color='black', linewidth=1)
plt.title('Oscilação de Tensão em Corrente Alternada (Projeto A3)')
plt.xlabel('Tempo (t) [s]')
plt.ylabel('Tensão (V) [Volts]')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

###**🏗️ Exercício 6: Função Quadrática — Engenharia Civil (Resistência de Vigas)**
Nesta modelagem, tanto a altura $h$ quanto a distância horizontal $x$ são expressas em metros, considerando o vão da viga no intervalo de $0 \le x \le 10$. Com base nas propriedades das funções de segundo grau, determine a posição exata $x$ onde a viga atinge seu ponto mais alto e qual é o valor dessa altura máxima em relação ao nível do solo. Além disso, utilize a linguagem Python para plotar o perfil gráfico da estrutura, destacando visualmente o vértice da parábola que representa o ponto de maior elevação.

**✅RESPOSTA**


✏️ **Resolução Analítica**:


<div align="justify">


Altura Máxima ($h_{max}$): Substituímos o valor de $x = 5$ na função original para encontrar a altura correspondente:
$$h(5) = -0,05(5^2) + 0,5(5) + 4$$$$h(5) = -1,25 + 2,5 + 4 = 5,25 \, m$$**Resultado final:** A viga atinge sua altura máxima de $5,25 \, m$ exatamente no centro da estrutura, na posição $x = 5 \, m$.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def perfil_viga(x):
    return -0.05 * x**2 + 0.5 * x + 4

x_viga = np.linspace(0, 10, 100)
h_viga = perfil_viga(x_viga)

plt.figure(figsize=(10, 5))
plt.plot(x_viga, h_viga, color='brown', linewidth=3, label='Perfil da Viga (Parábola)')
plt.fill_between(x_viga, h_viga, color='gray', alpha=0.1)
x_max = 5
h_max = 5.25
plt.scatter(x_max, h_max, color='red', s=80, zorder=5, label=f'Altura Máxima: {h_max}m')
plt.title('Análise Estrutural: Perfil Geométrico da Viga (Projeto A3)', fontsize=14)
plt.xlabel('Distância Horizontal (x) [m]', fontsize=12)
plt.ylabel('Altura em relação ao Solo (h) [m]', fontsize=12)
plt.ylim(0, 6)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='lower center')

plt.show()
```

###**⚙️ Exercício 7: Função Trigonométrica - Engenharia Mecânica (Movimento em uma Roda-Gigante)**

Uma roda-gigante tem raio de **10 metros** e seu centro está a **12 metros** do chão. Um passageiro embarca no ponto mais baixo da roda (ou seja, a **2 metros** do chão). A altura h(t) do passageiro em relação ao solo, em função do tempo t (em segundos), pode ser modelada por uma função trigonométrica.
Sabendo que a roda completa uma volta a cada 20 segundos:

Determine a função h(t).
Calcule a altura do passageiro após 5 segundos.
Faça um gráfico da função no intervalo de 0 a 40 segundos.





**✅ RESPOSTA**


✏️ **Resolução Analítica**:

<div align="justify">

<br> Construção da função:

Raio = **10**

Amplitude = **10**

Altura do centro = **12**

Deslocamento vertical = **12**

Período = **20** → **𝜋/10**

**Como começa no ponto mais baixo, usamos cosseno negativo:**

$$ℎ(𝑡)=12−10cos(𝜋/10𝑡)$$

**Altura em 𝑡 = 5**

$$h(5)=12−10cos(𝜋/10⋅5)$$

$$ℎ(5)=12−10cos(𝜋/2)$$

$$cos(𝜋2)=0$$

$$ℎ(5)=12−0=12$$

👉 **Altura = 12 metros**

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def h(t):
    return 12 - 10 * np.cos((np.pi/10) * t)

t = np.linspace(0, 40, 400)

altura = h(t)



plt.figure(figsize=(8,5))
plt.plot(t, altura, label='h(t) = 12 - 10cos(pi/10 * t)')
plt.axhline(12, linestyle='--', label='Altura do centro')
plt.title('Altura do passageiro na roda-gigante (Projeto A3)')
plt.xlabel('Tempo (s)')
plt.ylabel('Altura (m)')
plt.legend()
plt.grid()

plt.show()
```

 ### **🏗️ Exercício 8: Função Quadrática - Engenharia Civil (Variação do Nível da Água em Função do Tempo)**

 O nível da água em um porto varia ao longo do tempo devido à maré. Sabe-se que:

A altura média da água é de **6 metros**. A variação máxima (amplitude) é de **2 metros**. O ciclo completo da maré ocorre a cada 12 horas. Considerando que no instante **T=0** a maré está no nível médio e subindo:

Determine a função trigonométrica que modela a altura da água ao longo do tempo.
Calcule a altura da água após 3 horas.
Construa o gráfico da função no intervalo de 0 a 24 horas.

✅ **RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">

<br> **1. Construção da função :**

Amplitude = **2**

Deslocamento vertical = **6**

Período = **12**

$$𝜔=2𝜋/12=𝜋/6$$

Como começa no nível médio e subindo, usamos o seno:


$$ℎ(𝑡)=6+2sin⁡(𝜋/6𝑡)$$

📌 **2. Altura em T=3**

$$ℎ(3)=6+2sin⁡(𝜋/6⋅3)$$

$$ℎ(3)=6+2sin(𝜋/2)$$

$$sin(𝜋/2)=1$$

$$h(3)=6+2=8$$

👉 **Altura = 8 metros**

```python
import numpy as np
import matplotlib.pyplot as plt

def h(t):
    return 6 + 2 * np.sin((np.pi/6) * t)

t = np.linspace(0, 24, 400)

altura = h(t)



plt.figure(figsize=(8,5))
plt.plot(t, altura, label='h(t) = 6 + 2sen(pi/6 * t)')
plt.axhline(6, linestyle='--', label='Nível médio')
plt.title('Variação do nível da água (Projeto A3)')
plt.xlabel('Tempo (horas)')
plt.ylabel('Altura (m)')
plt.legend()
plt.grid()

plt.show()
```

### **⚡ Exercício 9: Função Linear — Engenharia Elétrica (Variação Térmica)**

<div align="justify">

Dados laboratoriais confirmaram que, para este material específico, a resistência cresce a uma taxa constante de $0,16 \, \Omega$ para cada grau Celsius de aumento na temperatura. Com base nessas informações, determine a lei de formação da função linear $R(T)$ que descreve o comportamento da resistência em relação à temperatura ambiente. Utilizando essa função, calcule qual será a resistência elétrica do filamento quando o sistema atingir uma temperatura de operação de $75°C$. Por fim, utilize a biblioteca Matplotlib do Python para gerar um gráfico que ilustre essa variação térmica no intervalo de $20°C$ a $100°C$, destacando o ponto correspondente à temperatura de operação calculada.

✅ **RESPOSTA**





✏️ **Resolução Analítica**:

<div align="justify">


Podemos escrever a função considerando o deslocamento a partir dos $20°C$:$$R(T) = 0,16 \cdot (T - 20) + 40$$Ao distribuirmos o valor $0,16$ e simplificarmos a expressão, chegamos à função final:$$R(T) = 0,16T - 3,2 + 40 \implies R(T) = 0,16T + 36,8$$


Cálculo da Resistência na Operação ($75°C$): Agora, basta substituirmos o valor da temperatura de operação na função encontrada:$$R(75) = 0,16 \cdot (75) + 36,8$$$$R(75) = 12 + 36,8 = 48,8 \, \Omega$$

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def resistencia(T):
    return 0.16 * T + 36.8

temp = np.linspace(20, 100, 100)
res_valores = resistencia(temp)


plt.figure(figsize=(9, 4))
plt.plot(temp, res_valores, color='orange', linewidth=2, label='Resistência $R(T)$')

plt.scatter(75, 48.8, color='black', zorder=5, label='Ponto Crítico (75°C, 48.8$\\Omega$)')

plt.title('Variação da Resistência Térmica no Condutor (Projeto A3)')
plt.xlabel('Temperatura (T) [°C]')
plt.ylabel('Resistência (R) [$\\Omega$]')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

###**⚡ Exercício 10: Função Exponencial — Engenharia Elétrica (Descarga de Capacitores)**




<div align="justify">


Determine qual será a tensão residual no circuito após $2$ segundos da queda de energia. Além disso, utilize a linguagem Python para plotar a curva de descarga desse componente nos primeiros $10$ segundos, permitindo visualizar o comportamento do decaimento da energia no sistema.

✅ **RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">

Cálculo da Tensão para $t = 2$ segundos: Substituímos o valor de $t$ na fórmula:$$V(2) = 12 \cdot e^{-0,5 \cdot 2}$$$$V(2) = 12 \cdot e^{-1}$$         

  Resolução Numérica:   $$  e^{-1} =\frac{1}{e} \approx 0,3678$$ $$V(2) = 12 \cdot 0,3678 \approx 4,41 \, V$$

**Resultado final**: Após $2$ segundos sem energia externa, a tensão no capacitor caiu de $12 \, V$ para aproximadamente $4,41 \, V$.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def descarga_capacitor(t):
    return 12 * np.exp(-0.5 * t)

tempo_descarga = np.linspace(0, 10, 100)
tensao_valores = descarga_capacitor(tempo_descarga)

plt.figure(figsize=(9, 4))
plt.plot(tempo_descarga, tensao_valores, color='purple', linewidth=2.5, label='Curva de Descarga ($V$)')

plt.scatter(2, 4.41, color='red', zorder=5, label='Tensão em 2s (4.41V)')

plt.title('Decaimento Exponencial de Tensão no Capacitor (Projeto A3)')
plt.xlabel('Tempo (t) [segundos]')
plt.ylabel('Tensão (V) [Volts]')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.show()
```

<div align="justify">


Um engenheiro de som está testando um novo pré-amplificador onde a potência de entrada é de $0,5 \, W$ (Watts) e a potência de saída atinge $50 \, W$. Determine o ganho de potência desse sistema em decibéis. Além disso, utilize o Python para plotar o gráfico do ganho $G$ em função da potência de saída $P_s$, variando de $1 \, W$ até $100 \, W$, mantendo a entrada fixa em $0,5 \, W$.

✅ **RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">


Cálculo do Ganho ($G$): Substituímos os valores na fórmula do enunciado:$$G = 10 \cdot \log_{10}\left(\frac{50}{0,5}\right)$$$$G = 10 \cdot \log_{10}(100)$$


**Resolução do Logaritmo:**  $$100 = 10^2, o \log_{10}(100) = 2$$ $$G = 10 \cdot 2 = 20 \, dB$$

**Resultado final**: O amplificador proporciona um ganho de $20 \, dB$ sobre o sinal original.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_ganho(Ps, Pe=0.5):
    return 10 * np.log10(Ps / Pe)

potencia_saida = np.linspace(1, 100, 500)
ganho_db = calcular_ganho(potencia_saida)

plt.figure(figsize=(9, 4))
plt.plot(potencia_saida, ganho_db, color='forestgreen', linewidth=2, label='Curva de Ganho (dB)')

plt.scatter(50, 20, color='red', zorder=5, label='Ponto de Teste (50W, 20dB)')

plt.title('Ganho de Potência em Escala Logarítmica (Projeto A3)')
plt.xlabel('Potência de Saída (Ps) [Watts]')
plt.ylabel('Ganho (G) [dB]')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

###**💻 Exercício 13: Função Racional — Engenharia de Software (Lei de Amdahl Simplificada)**




<div align="justify">

Nesta função, o termo $\frac{40}{n}$ representa a parte do código que pode ser executada em paralelo, enquanto o valor constante $10$ representa a parte sequencial do código que não pode ser dividida entre os núcleos. Com base nesse modelo, calcule qual será o tempo de resposta do sistema ao utilizar exatamente $4$ núcleos de processamento. Além disso, determine o limite teórico de execução do sistema, ou seja, o que acontece com o tempo $T$ quando o número de núcleos cresce infinitamente ($n \to \infty$). Por fim, utilize o Python para plotar a curva de desempenho para $n$ variando de $1$ a $20$ núcleos.

✅ **RESPOSTA**

✏️ **Resolução Analítica**:


<div align="justify">

Cálculo para 4 Núcleos ($n = 4$): Substituímos o valor de $n$ na função racional $T(n) = \frac{40}{n} + 10$:$$T(4) = \frac{40}{4} + 10$$$$T(4) = 10 + 10$$$$T(4) = 20 \, \text{segundos}$$Análise do Limite Teórico ($n \to \infty$):Para descobrir o que acontece quando adicionamos "infinitos" processadores, calculamos o limite da função quando $n$ cresce indefinidamente:$$\lim_{n \to \infty} \left( \frac{40}{n} + 10 \right)$$



À medida que o denominador ($n$) fica cada vez maior, a fração $\frac{40}{n}$ torna-se cada vez menor, aproximando-se de zero. Portanto: $0 + 10 = 10$.

**Resultado final**: Com 4 núcleos, o tempo de execução cai pela metade em relação a 1 núcleo (que seria 50s), resultando em 20 segundos. O limite teórico de execução é de 10 segundos. Isso significa que, não importa quanto dinheiro seja gasto em hardware extra, o sistema nunca será mais rápido que 10s, pois essa é a parte do código que não pode ser paralelizada.

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt


def tempo_execucao(n):
    return (40 / n) + 10


nucleos = np.arange(1, 21)
tempos = tempo_execucao(nucleos)


plt.figure(figsize=(9, 4))
plt.plot(nucleos, tempos, 'o-', color='darkorange', label='Tempo de Execução')


plt.axhline(y=10, color='blue', linestyle='--', label='Limite Teórico (10s)')


plt.scatter(4, 20, color='black', zorder=5)
plt.annotate(' 20s (4 núcleos)', (4, 21))

plt.title('Análise de Escalabilidade: Lei de Amdahl em Software (Projeto A3)')
plt.xlabel('Número de Núcleos (n)')
plt.ylabel('Tempo de Resposta (segundos)')
plt.xticks(range(1, 21))
plt.grid(True, alpha=0.2)
plt.legend()
plt.show()
```

###**🏗️ Exercício 14: Função Racional — Engenharia Civil (Distribuição de Pressão em Fundações)**


<div align="justify">

A pressão $P$ exercida no solo por uma fundação quadrada pode ser modelada pela função $P(n) = \frac{500}{n} + 20$, onde $n$ representa a área da base da fundação em metros quadrados ($m^2$) e o termo constante $20$ representa a sobrecarga permanente do solo. Com base nesse modelo, calcule qual será a pressão exercida pelo solo ao utilizar uma base de $n = 5$ $m^2$. Além disso, determine o comportamento da pressão $P$ quando a área da base cresce indefinidamente ($n \to \infty$). Por fim, utilize o Python para plotar a curva de pressão para $n$ variando de $1$ a $25$ $m^2$.

✅ **RESPOSTA**

✏️ **Resolução Analítica**:

<div align="justify">




Cálculo para uma base de 5 $m^2$ ($n = 5$): Substituímos o valor de $n$ na função racional $P(n) = \frac{500}{n} + 20$:$$P(5) = \frac{500}{5} + 20$$$$P(5) = 100 + 20$$$$P(5) = 120 \text{ kPa}$$Limite teórico ($n \to \infty$):$$\lim_{n \to \infty} \left( \frac{500}{n} + 20 \right) = 0 + 20 = 20 \text{ kPa}$$

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def pressao_fundacao(n):
    return (500 / n) + 20

area = np.linspace(1, 25, 100)
pressao = pressao_fundacao(area)

plt.figure(figsize=(9, 4))
plt.plot(area, pressao, color='brown', linewidth=2.5, label='Curva de Pressão $P(n)$')

plt.scatter(5, 120, color='red', zorder=5, label='Ponto de Teste (n=5, P=120)')

plt.axhline(y=20, color='gray', linestyle='--', label='Sobrecarga do Solo (20)')

plt.title('Análise de Pressão em Fundações (Projeto A3)')
plt.xlabel('Área da Base (n) [m²]')
plt.ylabel('Pressão (P)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

###**🔊Exercício 15: Função Logarítmica — Engenharia Acústica (Atenuação de Ruído em Materiais)**

<div align="justify">

O nível de intensidade sonora $L$ (em dB) após atravessar uma parede com espessura $x$ (em cm) de um material isolante é dado pela função $L(x) = 80 - 10 \cdot \log_{10}(x + 1)$. Sendo $80$ o nível de ruído inicial em dB, calcule o nível sonoro resultante após a passagem por um material com espessura de $x = 9$ cm. Além disso, determine a espessura necessária para que o nível de ruído atinja o limite de $40$ dB. Por fim, utilize o Python para plotar a curva de atenuação para $x$ variando de $0$ a $100$ cm.

✅ **RESPOSTA**


✏️ **Resolução Analítica**:

<div align="justify">





Cálculo para espessura de 9 cm ($x = 9$): Substituímos o valor de $x$ na função logarítmica
 $L(x) = 80 - 10 \cdot \log_{10}(x + 1)$:

 $$L(9) = 80 - 10 \cdot \log_{10}(9 + 1)$$$$
 L(9) = 80 - 10 \cdot \log_{10}(10)$$$$L(9) = 80 - 10(1) = 70 \text{ dB}$$

 Cálculo para atingir 40 dB ($L(x) = 40$):

 $$40 = 80 - 10 \cdot \log_{10}(x + 1)$$$$-40 = -10 \cdot \log_{10}(x + 1)
 $$$$4 = \log_{10}(x + 1)$$

 $$10^4 = x + 1 \implies x = 10.000 - 1 = 9.999 \text{ cm}$$

🐍 **Código Python**

```python
import numpy as np
import matplotlib.pyplot as plt

def atenuacao_sonora(x):
    return 80 - 10 * np.log10(x + 1)

espessura = np.linspace(0, 100, 500)
intensidade = atenuacao_sonora(espessura)

plt.figure(figsize=(9, 4))
plt.plot(espessura, intensidade, color='darkgreen', linewidth=2.5, label='Nível Sonoro L(x)')

plt.scatter(9, 70, color='red', zorder=5, label='Intensidade em 9cm (70dB)')

plt.axhline(y=40, color='gray', linestyle='--', label='Limite de 40dB')

plt.title('Atenuação Acústica: Eficiência do Material Isolante (Projeto A3)')
plt.xlabel('Espessura do Isolante (x) [cm]')
plt.ylabel('Intensidade Sonora (L) [dB]')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```
